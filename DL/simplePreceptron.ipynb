```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("Datasets/diabetes.csv")
X = df.drop("Outcome",axis=1)
Y = df["Outcome"]
sc = StandardScaler()
X_scale = sc.fit_transform(X)
X_train,x_test,Y_train,y_test = train_test_split(X_scale,Y,random_state=42)
w1 = np.random.randn(8)
b1 = 2

w2 = np.random.randn(8)
b2 = 1.3

w3 = np.random.randn(8)
b3 = 1.22

w4 = np.random.randn(3)
b4 = np.random.random()

w5 = np.random.randn(5)
b5 = np.random.random()

w6 = np.random.randn(3) 
b6 = np.random.random()

def sigmoid(z):
        return 1 / (1 + np.exp(-z))
y_pred = []
for i in range(576):
    i1 = X_train[i]
    h1p1 = np.dot(i1,w1)+b1
    
    i2 = X_train[i]
    h1p2 = np.dot(i2,w2)+b2
    
    i3 = X_train[i]
    
    h1p3 = np.dot(i3,w3)+b3

    h1 = [h1p1,h1p2,h1p3]

    h2 = []
    for i in range(5):
        h2p = np.dot(w4,h1) +b4 
        h2p = max(0,h2p)
        h2.append(h2p)
    h3 = []
    for i in range(3):
        h3p = np.dot(w5,h2) +b5 
        h3p = max(0,h3p)
        h3.append(h3p)
    
    z = np.dot(h3,w6)+b6
    y_pred.append(sigmoid(z))
```

```python
print(len(X_train))
print(len(Y_train))
print(len(x_test))
print(len(y_test))
print(len(y_pred))
```

```python
y_pred = [0 if i<0.5 else 1 for i in y_pred]
```
