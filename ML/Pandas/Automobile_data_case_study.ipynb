```python
#separate numerical type column and object type column and store in new dataset
df_num=df.select_dtypes(["int64","float64"])    #hold to int and float type data 
df_cat=df.select_dtypes(object)     #hold object type data
```

```python
df_num.info()
```

```python
df_cat.info()
```

```python
#change the object type data into numerical type then use LabelEncoder class
from sklearn.preprocessing import LabelEncoder
```

```python
df_cat.head()
```

```python
for col in df_cat :
    le=LabelEncoder() #create the object of LabelEncoder class
    df_cat[col]=le.fit_transform(df_cat[col])
```

```python
df_cat.head()
```

```python
#concatenation of both dataset df_num and df_cat and hold new dataset df_new
df_new=pd.concat([df_num,df_cat],axis=1) #axis=1 column wise
df_new.head()
```

```python
df_new.columns
```

```python
#performing the model with the help of linear regression
#we want to predict the car's price which are contineous numeric value
#apply multiple linear regression because many input means many independent
#variables or features and one target
```

```python
#select independent means input variables(X) and target means dependent 
#variable(Y)
X=df_new.drop("price",axis=1) #apart from price, all columns hold in X variable
Y=df_new["price"] #output variable
```

```python
X.columns
```

```python
#train test split 70%-30%
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,
                                              random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
#Apply StandardScaler on X_train and X_test
from sklearn.preprocessing import StandardScaler
#create the object of StandardScaler class
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
X_train
```

```python
#PERFORM MODEL WITH LINEARREGRESSION
from sklearn.linear_model import LinearRegression
#create object of LinearRegression class
lr=LinearRegression()
```

```python
#train the model with 70% data
lr.fit(X_train,Y_train)
```

```python
#check the training score on training data ,
#use inbuilt method score() of LinearRegression class
lr.score(X_train,Y_train)
```

```python
#check slope or coefficient of each input , use inbuilt method coef_
lr.coef_
```

```python
X.columns
```

```python
#check testing score with 30% data
lr.score(X_test,Y_test)
```

```python
#training score =0.85  and testing score =0.79 
#here model is overfit because training score >testing score
```

```python
#create a dictionary whoch hold input column and coefficient of input
dict={'features':X.columns,'Slope':lr.coef_}
df1=pd.DataFrame(dict)
df1
```

```python
#Apply Regularisation . bcoz model is overfit 
#1. Lasso Regularisation (L1)
#In Regularisation , we add some external error during training time 
#in Lasso : formula : RSS+lambda*sum(absolute(All slope))
#lambda means hyperparameter 
#in Lasso , range of lambda = 200 to 1000
```

```python
#call inbuilt class Lasso  
from sklearn.linear_model import Lasso
```

```python
#How to select the right value of lambda /alpha in lasso , use simple
#technique 
for i in range(200,1001,50): #start=200 stop=1001-1=1000 step=+50
    #create the object of Lasso class , passing the value of lambda
    l1=Lasso(i) #first i=200 , next i=250 ......
    #automatic write formula RSS+i*sum(abs(all coef))
    #train the model with 70% data using Lasso class , use fit() inbuilt
    #method of Lasso class
    l1.fit(X_train,Y_train)
    print("Value of Lambda : ",i)
    #check training score 
    print("Training Score : ",l1.score(X_train,Y_train))
    #check testing score
    print("Testing Score : ",l1.score(X_test,Y_test))
# l1=Lasso(1) 
# l1.fit(X_train,Y_train)
# max = l1.score(X_test,Y_test)
# print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
# for i in range(1,1001,1): #start=200 stop=1001-1=1000 step=+50
#     #create the object of Lasso class , passing the value of lambda
#     l1=Lasso(i) #first i=200 , next i=250 ......
#     #automatic write formula RSS+i*sum(abs(all coef))
#     #train the model with 70% data using Lasso class , use fit() inbuilt
#     #method of Lasso class
#     l1.fit(X_train,Y_train)
#     #check training score 
#     if l1.score(X_test,Y_test)>max:
#         print("Value of Lambda : ",i)
#         max = l1.score(X_test,Y_test)
#         print("Training Score : ",l1.score(X_train,Y_train))
#     #check testing score
#         print("Testing Score : ",l1.score(X_test,Y_test))  
```

```python
#Right value of Lambda in the case of Lasso =200 
#score got 81% and model is not overfit because training score=84%
#check coefficient /slope of each input/features after apply lasso
l1.coef_
```

```python
#To add new column of slope in existing dataframe df1 after apply Lasso
df1['Slope_Lasso']=l1.coef_
df1
```

```python
#Y_pred=m1*X1+m2*X2+..........mn*Xn+C
#Lasso use to select important features from given dataset 
#Lasso is a features selection method.
```

```python
#Ridge Regularisation 
#call Ridge class inbuilt class
from sklearn.linear_model import Ridge
```

```python
#in ridge 
#RSS+lambda*sum(square(ALL Slope))
#range of lambda 0-50
for i in range(0,51): #start=0 stop=51-1=50 step=+1
    #create the object of Ridge means L2 and passing the value of lambda 
    l2=Ridge(i)
    #train the model with 70% data using Ridge class , use fit() inbuilt
    #method of Ridge class
    l2.fit(X_train,Y_train)
    print("Value of Lambda : ",i)
    #check training score 
    print("Training Score : ",l2.score(X_train,Y_train))
    #check testing score
    print("Testing Score : ",l2.score(X_test,Y_test))
     
```

```python
#right value of lambda in Ridge :  12
#got testing score =81% and training score=84%  means model is not overfit
```

```python
#check coefficient after apply Ridge
l2.coef_
```

```python
df1['Slope_Ridge']=l2.coef_
df1
```

```python
#Cross validation : - 
#score are same in both regularisation (Lasso and  Ridge) 0.81
#Select any one regularisation method : use cross validation
#inbuilt class for cross validation :  cross_val_score
from sklearn.model_selection import cross_val_score
```

```python
#In Lasso : right value of lambda =200
#create object of Lasso class and passing the value of lambda =200
l1=Lasso(200)
```

```python
#Apply StandardScalar on input X
#create the object of StandardScalar class

ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
#Create object of cross_val_score class(for Lasso)
l1_cross=cross_val_score(l1,X,Y,cv=4)
#here X input and Y output 
#parameter cv means divided the given dataset into no. of parts
#here cv=4 means 4 parts 
```

```python

l1_cross
```

```python
#Find the average or means of all scores 
np.mean(l1_cross)
```
