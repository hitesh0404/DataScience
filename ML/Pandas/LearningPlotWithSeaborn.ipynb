```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 5, 3, 6, 4]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data points as markers
ax.plot(x, y, marker='o', linestyle='', color='blue', markersize=8)

# Annotate each data point with its value
for i, j in zip(x, y):
    ax.annotate(f'({i},{j})', xy=(i, j), xytext=(5, 5), textcoords='offset points',
                ha='left', va='bottom', fontsize=9, color='black')

# Add labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Data Points with Values")
ax.grid(True)

# Display the plot
plt.show()
```

```python
def hr_cat(data):
    if data<60 :
        return "Low"
    elif data >=60 and data<=70:
        return "Mid"
    else:
        return "High"

df["HourlyRate_grp"] = pd.Series(map(hr_cat,df["HourlyRate"]))
```
