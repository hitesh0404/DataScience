```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier,plot_tree
df = pd.read_csv("../../DataSets/logistic_regression_1000.csv")
```

```python
X = df.drop("Pass_Fail",axis=1)
Y = df["Pass_Fail"]
X_train,X_test,y_train,y_test = train_test_split(X,Y,train_size=0.8,random_state=42)
```

```python
deeptree = DecisionTreeClassifier()
deeptree.fit(X_train,y_train)
y_pred = deeptree.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
pruned = DecisionTreeClassifier(criterion='gini',max_depth=6,min_samples_leaf=2,min_samples_split=8,max_features="sqrt")
pruned.fit(X_train,y_train)
y_pred = pruned.predict(X_test)
accuracy_score(y_test,y_pred)
```
