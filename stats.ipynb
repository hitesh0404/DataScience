```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Automobile_data.csv")
```

```python
df.head()
```

```python
sns.scatterplot(df,x="horsepower",y="price")
plt.show()
```

```python
ss = StandardScaler()
```

```python
data = df[["horsepower","price"]]
```

```python
data.loc[(130,131),"horsepower"] = '0'
data.loc[:,"horsepower"] = data.horsepower.astype("int64")
```

```python
data.loc[(130,131),"horsepower"] = data.horsepower.median()
```

```python
data
```

```python
df_ss = ss.fit_transform(data)
```

```python
df_ss = pd.DataFrame(df_ss)
```

```python
ss.inverse_transform(df_ss)
```

```python
sns.scatterplot(df_ss,x=0,y=1)
plt.show()
```
